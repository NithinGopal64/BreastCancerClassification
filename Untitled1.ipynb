```python
from __future__ import print_function

import numpy as np

import argparse
import torch
import torch.utils.data as data_utils
import torch.optim as optim
from torch.autograd import Variable

from dataloader import MnistBags
from model import Attention, GatedAttention
from patch_data import PatchMethod



# Training settings
# parser = argparse.ArgumentParser(description="Breakthis data_mynet")
# parser.add_argument("--zoom", help="zoom_level", default=400)
zoom = 40
# parser.add_argument(
#     "--epochs",
#     type=int,
#     default=1,
#     metavar="N",
#     help="number of epochs to train (default: 10)",
# )
epochs = 1
# parser.add_argument(
#     "--seed", type=int, default=1, metavar="S", help="random seed (default: 1)"
# )
seed = 1
# parser.add_argument(
#     "--no-cuda", action="store_true", default=False, help="disables CUDA training"
# )
cuda = True
# parser.add_argument(
#     "--lr",
#     type=float,
#     default=0.0001,
#     metavar="LR",
#     help="learning rate (default: 0.01)",
# )
lr = 0.0001
# parser.add_argument(
#     "--reg", type=float, default=10e-5, metavar="R", help="weight decay"
# )
reg = 10e-5
# args = parser.parse_args()
# args.cuda = not args.no_cuda and torch.cuda.is_available()

torch.manual_seed(seed)
if cuda:
    torch.cuda.manual_seed(seed)
    print("\nGPU is ON!")

print("Load Train and Test Set")
loader_kwargs = {"num_workers": 1, "pin_memory": True} if cuda else {}
data_path_train = "data_breakhis/40/train"
data_path_test = "data_breakhis/40/test"
data = PatchMethod(root=data_path_train)
val_data = PatchMethod(root=data_path_test)
train_loader = data_utils.DataLoader(data, shuffle=True, batch_size=1)

test_loader = data_utils.DataLoader(val_data, shuffle=False, batch_size=1)

print("Init Model")
model = Attention()
if cuda:
    model.cuda()

optimizer = optim.Adam(
    model.parameters(), lr=lr, betas=(0.9, 0.999), weight_decay=reg
)


def train(epoch):
    model.train()
    train_loss = 0.0
    train_error = 0.0
    le = len(train_loader)
    for batch_idx, (data, label) in enumerate(train_loader):
        bag_label = label[0]
        if batch_idx % 100 == 99:
            print(f"{batch_idx+1}/{le}")
        if cuda:
            data, bag_label = data.cuda(), bag_label.cuda()
        data, bag_label = Variable(data), Variable(bag_label)
        # reset gradients
        optimizer.zero_grad()
        # calculate loss and metrics
        loss, _ = model.calculate_objective(data, bag_label)
        train_loss += loss.data[0]
        error, _ = model.calculate_classification_error(data, bag_label)
        train_error += error
        # backward pass
        loss.backward()
        # step
        optimizer.step()

    # calculate loss and error for epoch
    train_loss /= len(train_loader)
    train_error /= len(train_loader)

    print(
        "Epoch: {}, Loss: {:.4f}, Train error: {:.4f}".format(
            epoch, train_loss.cpu().numpy()[0], train_error
        )
    )


def test():
    model.eval()
    test_loss = 0.0
    test_error = 0.0
    for batch_idx, (data, label) in enumerate(test_loader):
        bag_label = label[0]
        instance_labels = label[0]
        if cuda:
            data, bag_label = data.cuda(), bag_label.cuda()
        data, bag_label = Variable(data), Variable(bag_label)
        loss, attention_weights = model.calculate_objective(data, bag_label)
        test_loss += loss.data[0]
        error, predicted_label = model.calculate_classification_error(data, bag_label)
        test_error += error
    test_error /= len(test_loader)
    test_loss /= len(test_loader)

    print(
        "\nTest Set, Loss: {:.4f}, Test error: {:.4f}".format(
            test_loss.cpu().numpy()[0], test_error
        )
    )

train(1)
# if __name__ == "__main__":
#     print("Start Training")
#     for epoch in range(1, args.epochs + 1):
#         train(epoch)
#     print("Start Testing")
#     test()
```
